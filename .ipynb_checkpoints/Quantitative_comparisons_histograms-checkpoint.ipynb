```python
import pandas as pd
```

```python
mens_rowing = pd.DataFrame({"Height":[198.0,178.0,183.0,188.0,187.0],
                           "Weight":[90.0,71.0,70.0,73.0,90.0]})
```

```python
mens_gymnastics = pd.DataFrame({"Height":[150.0,169.0,173.0,158.0,187.0],
                           "Weight":[90.0,71.0,70.0,73.0,90.0]})
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax=plt.subplots()
ax.hist(mens_rowing["Height"])
ax.hist(mens_gymnastics["Height"])
ax.set_xlabel("Heghts (cm)")
ax.set_ylabel("# observations")
plt.show()
```

```python
fig, ax=plt.subplots()
ax.hist(mens_rowing["Height"], label="Rowing", bins=5, histtype="step")
ax.hist(mens_gymnastics["Height"], label="Gymnastics", bins=5, histtype="step")
ax.set_xlabel("Heghts (cm)")
ax.set_ylabel("# observations")
ax.legend()
plt.show()
```
