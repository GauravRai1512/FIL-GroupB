```python
import matplotlib.pyplot as plt
```

```python
values =[0,0.6,1.4,1.6,2.2,2.4,2.6,2.9,3.5,3.2,4.5,4.6,5.9]

plt.hist(values, bins=5)
plt.show()
help(plt.hist)
plt.clf() #cleans it up
# here bins means how many histogram will be there like bins=5 means x axis 6/5=1.2, means 5 beans will be there with size 1.2
```
