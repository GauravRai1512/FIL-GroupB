```python
import pandas as pd
```

```python
medals = pd.DataFrame({'Countries' : ['United States','Germany','Great Britain','Russia','China','France','Australia','Italy','Canada','Japan'],
                      'Gold':[137,47,64,50,44,20,23,8,4,17],
                      'Silver':[52,43,55,28,30,55,23,38,4,13],
                      'Bronze' : [67,67,26,35,35,21,25,24,61,34]})
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

# Plot a bar-chart of gold medals as a function of country
ax.bar(medals["Countries"],medals["Gold"])

# Set the x-axis tick labels to the country names
ax.set_xticklabels(medals["Countries"], rotation=90)

# Set the y-axis label
ax.set_ylabel("Number of medals")

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals['Countries'],medals['Gold'],label='Gold')
ax.bar(medals['Countries'],medals['Silver'],bottom=medals['Gold'],label='Silver')
ax.bar(medals['Countries'],medals['Bronze'],bottom=medals['Gold'] + medals['Silver'],label='Bronze')
ax.set_xticklabels(medals['Countries'], rotation=90)
ax.legend()
plt.show()
```
